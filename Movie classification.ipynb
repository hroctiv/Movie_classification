```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import operator
import langdetect
```

```python
dfFull = pd.read_csv("movies_genres.csv",delimiter='\t')
dfFull.sample(10)
```

```python
df_genres = dfFull.drop(["plot", "title"], axis=1)
counts = []
categories = df_genres.columns
for cat in categories:
    counts.append((cat, df_genres[cat].sum()))
df_stats = pd.DataFrame(sorted(counts, key=operator.itemgetter(1)), columns=["genre", "count"])
df_stats
```

```python
#we'll remove Lifestyle since there are no occurences
df_genres = df_genres.drop(["Lifestyle"],axis =1)
```

```python
dfFull["plot_lang"] = dfFull.apply(lambda row: langdetect.detect(row['plot']),axis=1)
dfFull["plot_lang"].value_counts()
```
